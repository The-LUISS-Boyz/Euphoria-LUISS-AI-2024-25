```python
def convert_to_timestamp(value):
    spl = value.split(',')

    # Try to convert each element in the split list to a timestamp
    for i in spl:
        try:
            # Try converting the value to a float and then to a timestamp
            timestamp = float(i)
            return datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
        except ValueError:
            # If it fails to convert, return 'non'
            continue

    return 'non'  # Return 'non' if none of the values were successfully converted

# Apply the function to the 'referral_friends' column and store the result in 'creation_time'
data['creation_time'] = data['referral_friends'].apply(convert_to_timestamp)
```

## Numeric-Values Filtering Graph

In this work, we shed light on how pre-processing will change numeric features of the data set. More precisely, most numeric data requires scaling or normalization to make data consistent and suitable for further analytical techniques.

To compare these changes, we visualized the distributions of numeric features before and after preprocessing. We created side-by-side histograms for each feature to highlight how scale, range, and distribution change with the preprocessing steps. This comparison helps identify the impact of transformations, ensuring the processed data aligns with analytical objectives.

Visualization of these changes, we informed the preprocessing outcome and verified some important changes regarding data so that our movement to other stages, preprocessing, and analysis or modeling, becomes quite smooth.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
raw_data = pd.read_csv('euphoria dataset.csv')  
processed_data = pd.read_csv('clean_data.csv')  

def plot_numeric_features(raw_df, processed_df, features):
    """
    Visualize all numeric features before and after preprocessing.

    Parameters:
    raw_df (DataFrame): Raw dataset.
    processed_df (DataFrame): Processed dataset.
    features (list): List of numeric features to visualize.
    """
    for feature in features:
        # Ensure feature exists in both datasets and is numeric
        if feature in processed_df.columns and processed_df[feature].dtype in ['int64', 'float64']:
            plt.figure(figsize=(12, 5))

            # Plot raw data distribution
            plt.subplot(1, 2, 1)
            raw_df[feature].dropna().plot(kind='hist', bins=30, alpha=0.7, title=f"Raw Data: {feature}")
            plt.xlabel(feature)
            plt.ylabel("Frequency")

            # Plot processed data distribution
            plt.subplot(1, 2, 2)
            processed_df[feature].dropna().plot(kind='hist', bins=30, alpha=0.7, title=f"Processed Data: {feature}")
            plt.xlabel(feature)
            plt.ylabel("Frequency")

            plt.tight_layout()
            plt.show()
        else:
            print(f"Skipping feature '{feature}': Not numeric in processed dataset.")

# Identify numeric features in the raw dataset
numeric_features_raw = [col for col in raw_data.columns if raw_data[col].dtype in ['int64', 'float64']]

# Visualize all numeric features
plot_numeric_features(raw_data, processed_data, numeric_features_raw)
```

## Steps for Identifying the Problem Type:
Examine the Target Variable:

Regression: If the target variable (e.g., happiness_index) is continuous (numerical), it is a regression problem.
Classification: If the target variable is categorical (e.g., happiness_metric with values like 'High', 'Medium', 'Low'), it is a classification problem.
Clustering: If the problem involves grouping data points into clusters without a clear target variable, it is a clustering problem.
Distribution of the Target Variable:

A continuous target (like happiness_index) suggests regression.
A categorical target (like happiness_metric) suggests classification.
Visual Analysis:

We can visualize the distribution of the target variable and decide if it's continuous or categorical.

Performance will be evaluated using appropriate metrics (e.g., accuracy, precision, recall, F1 score, or silhouette score for clustering).

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Initialize the models
lr = LinearRegression()
rf = RandomForestRegressor(random_state=42)
gb_regressor = GradientBoostingRegressor(random_state=42)

# Train the models
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
gb_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred_lr = lr.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_gb = gb_regressor.predict(X_test)

# Calculate MSE, R², and RMSE for each model
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)

mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)
rmse_gb = np.sqrt(mse_gb)

# Print the evaluation metrics for each model
print("Linear Regression - MSE: {:.4f}, R²: {:.4f}, RMSE: {:.4f}".format(mse_lr, r2_lr, rmse_lr))
print("Random Forest Regressor - MSE: {:.4f}, R²: {:.4f}, RMSE: {:.4f}".format(mse_rf, r2_rf, rmse_rf))
print("Gradient Boosting Regressor - MSE: {:.4f}, R²: {:.4f}, RMSE: {:.4f}".format(mse_gb, r2_gb, rmse_gb))
```

## Hyperparameter Tuning

- Hyperparameter tuning is a powerful method to improve model performance by adjusting model parameters to find the best possible configuration.

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from scipy.stats import uniform, randint
from sklearn.metrics import mean_squared_error

# Subset the data for faster tuning (optional)
X_sample, _, y_sample, _ = train_test_split(X_train, y_train, train_size=0.5, random_state=42)

# Define parameter distributions for RandomizedSearchCV
param_distributions = {
    'n_estimators': randint(100, 300),  # Number of trees
    'learning_rate': uniform(0.01, 0.1),  # Learning rate
    'max_depth': randint(3, 6),  # Depth of trees
    'min_samples_split': randint(2, 11),  # Minimum samples to split a node
    'subsample': uniform(0.8, 0.2)  # Fraction of samples for training each tree
}

# Initialize the Gradient Boosting Regressor
gb_regressor = GradientBoostingRegressor(random_state=42)

# Set up RandomizedSearchCV
randomized_search = RandomizedSearchCV(
    estimator=gb_regressor,
    param_distributions=param_distributions,
    n_iter=10,  # Limit to 50 random combinations
    cv=3,  # Use 3-fold cross-validation
    n_jobs=-1,  # Utilize all available CPU cores
    scoring='neg_mean_squared_error',
    random_state=42,
    verbose=1
)

# Fit the RandomizedSearchCV to the subset of the training data
randomized_search.fit(X_sample, y_sample)

# Get the best parameters
best_params = randomized_search.best_params_
print(f"Best Hyperparameters: {best_params}")

# Evaluate the best model on the test set
best_model = randomized_search.best_estimator_
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (Best Model): {mse}")


```
